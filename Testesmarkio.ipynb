```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
plt.style.use('ggplot')
```

```python
%matplotlib notebook
sns.distplot(df['ORÇAMENTO.INICIAL'],color='blue', kde=False)
plt.title('Distribuição da coluna ORÇAMENTO.INICIAL')
```

```python
%matplotlib notebook
sns.distplot(df['ORÇAMENTO.REALIZADO'],color='green', kde=False)
plt.title('Distribuição da coluna ORÇAMENTO.REALIZADO')
```

```python
df[['NOME.ÓRGÃO.SUPERIOR','ORÇAMENTO.INICIAL','ORÇAMENTO.REALIZADO']].corr()
```

```python
correlacoes = df[['NOME.ÓRGÃO.SUPERIOR','ORÇAMENTO.INICIAL','ORÇAMENTO.REALIZADO']].corr()
```

```python
%matplotlib notebook
sns.heatmap(correlacoes, annot=True, cmap="YlGnBu")
```
